```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 500)
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

from gen_figs import load_results
from utils.threshold_utils import get_log_lrPlus
```

```python
results, result_paths = load_results("MSH2_Jia_2021",
    ["/data/dzeiberg/mave_calibration/sc/arion/projects/pejaverlab/users/zeibed01/mave_calibration/results_10_19_24/"],
    includes_synonymous=False)
```

```python
prior = np.load("/data/dzeiberg/mave_calibration/minerva/figs_10_19_24/MSH2_Jia_2021/priors.npy")
```

```python
df = pd.read_csv("/data/dzeiberg/IGVF-cvfg-pillar-project/Pillar_project_data_files/individual_datasets/MSH2_Jia_2021.csv")
```

```python
rng = np.arange(df.auth_reported_score.min(), df.auth_reported_score.max(), .01)
log_lrs = np.stack([get_log_lrPlus(rng,1,result) for result in results])
```

```python
fig, ax = plt.subplots()
ax.plot(rng, log_lrs.mean(0))
ax.fill_between(rng, *np.percentile(log_lrs, [2.5, 97.5], axis=0), alpha=.5)
ax.set_xlabel("Assay Score")
ax.set_ylabel(r"$ \log LR^+$")
```
