```python
dataset_path = Path("/data/dzeiberg/mave_calibration/processed_datasets/")
```

```python
datasets = {file.stem.split("_pipeline")[0] : pd.read_json(file) for file in dataset_path.glob("*.json")}
```

```python
counts = {dataset_config[name]['curation_name'] if name in dataset_config and "curation_name" in dataset_config[name] else name : {"P/LP" : df.labels.apply(lambda li: "P/LP" in li).sum(),
                  "B/LB" : df.labels.apply(lambda li: "B/LB" in li).sum(),
                  "gnomAD": df.labels.apply(lambda li: "gnomAD" in li).sum(),
                  "synonymous": df.labels.apply(lambda li: "synonymous" in li).sum(),
                  "VUS" : df.labels.apply(lambda li: "VUS" in li).sum(),} for name, df in datasets.items()}
```
