```python
from utils.evidence_framework import get_tavtigian_constant
import numpy as np
from tqdm.autonotebook import tqdm
import matplotlib.pyplot as plt
```

```python
priors = np.arange(.01, .2,.01)
constant = np.array([get_tavtigian_constant(prior) for prior in tqdm(priors)])
```

```python
fig,(ax1,ax2) = plt.subplots(2,1,figsize=(10,10))
ax1.plot(priors, (1/8) * np.log(constant))
ax1.set_xlabel('Prior')
secax = ax1.secondary_yaxis('right', functions=(lambda x: np.exp(x), lambda x: np.log(x)))
# ax1.scatter([.0441],[np.log(2.406)])
ax1.set_ylabel("Log LR+ Pathogenic Supporting")
secax.set_ylabel("LR+ Pathogenic Supporting")

ax2.plot(priors, (-1/8) * np.log(constant))
ax2.set_xlabel('Prior')
secax2 = ax2.secondary_yaxis('right', functions=(lambda x: np.exp(x), lambda x: np.log(x)))
# ax2.scatter([.0441],[np.log(.4156)])
ax2.set_ylabel("Log LR+ Benign Supporting")
secax2.set_ylabel("LR+ Benign Supporting")
```
