```python
import pandas as pd
from pathlib import Path
import yaml
```

```python
dataset_path = Path("/data/dzeiberg/mave_calibration/processed_datasets/")
```

```python
datasets = {file.stem.split("_pipeline")[0] : pd.read_json(file) for file in dataset_path.glob("*.json")}
```

```python
with open("/home/dzeiberg/mave_results/datasets.yaml") as f:
    dataset_config = yaml.safe_load(f)
```

```python
counts = {i : \
    {"gene": dataset_config[name]['gene'],
    "experiment": dataset_config[name]['experiment'],
    "reference": "\cite{" + dataset_config[name]['reference']  + "}",
    "P/LP" : df.labels.apply(lambda li: "P/LP" in li).sum(),
    "B/LB" : df.labels.apply(lambda li: "B/LB" in li).sum(),
    "gnomAD": df.labels.apply(lambda li: "gnomAD" in li).sum(),
    "synonymous": df.labels.apply(lambda li: "synonymous" in li).sum(),
    "VUS" : df.labels.apply(lambda li: "VUS" in li).sum(),
    'missense': df.labels.apply(lambda li: "synonymous" not in li).sum(),
    } for i,(name, df) in enumerate(datasets.items())}
```

```python
counts_df = pd.DataFrame(counts).T
```
